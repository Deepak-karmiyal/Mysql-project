```python
import os

import mysql.connector
#from mysql import connector
from dotenv import load_dotenv
import pandas as pd
```

```python
load_dotenv()

user = os.getenv('USER')
password = os.getenv('PASSWORD')
host = os.getenv('HOST')
database = 'SwiftMarket'
```

```python
connection =  mysql.connector.connect( 
                                    user=user,
                                    password=password,
                                    host=host,
                                    database=database
                                    )
cursor = connection.cursor()
```

```python
connection.is_connected()
```

```python
query = '''show tables'''
cursor.execute(query)
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
query = '''select * from salesdata limit 5;'''
cursor.execute(query)
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
cursor.column_names
```

```python
pd.DataFrame(data=rows,columns=cursor.column_names)
```

```python
def mysqlquery (query):
    cursor.execute(query)
    rows = cursor.fetchall()    
    return pd.DataFrame(data=rows,columns=cursor.column_names)


```

```python
query = '''select * from customers limit 5'''
mysqlquery(query)
```

```python
query = '''select s.saledate,c.firstname,s.totalprice
from salesdata s
join customers c
on s.customerid = c.customerid
order by s.saledate;'''
mysqlquery(query)
```

```python
query = '''select * from subcategories'''
mysqlquery(query)
```
